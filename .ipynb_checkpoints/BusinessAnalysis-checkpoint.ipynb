### Text Preprocessing puntuation removal

```python
def clean_punctuation(text):
    text_lower = str(text).lower()
    text_np = text_lower.translate(str.maketrans('', '', string.punctuation))
    return text_np
text_collection_np = list(map(clean_punctuation,text_collection))
```

### Text preprocessing stop words elimination

```python
def clean_stop(text):
    stop_words = stopwords.words('english')
    word_tokens = word_tokenize(text)
    clean_text = [w for w in word_tokens if not w in stop_words]
    return clean_text
text_collection_rstop = list(map(clean_stop,text_collection_np)) 
```

### Text preprocessing stemming
For my particular application with very similar words I will use stemming to create the word normalization. Reasons, monotone use of words and short amount of word collection. 

```python
def tokenizer_stemmer(word_array):
    stemmer = WordNetLemmatizer()
    word_stem = []
    for word in word_array:
        word_stem.append(stemmer.lemmatize(word))
    return word_stem
text_stem = list(map(tokenizer_stemmer,text_collection_rstop))
```

```python
text_stem
```

```python
# Remove numbers
text_stem = [[token for token in text if not token.isdigit()] for text in text_stem]
```

```python
# Remove words with only 1 character
text_stem = [[token for token in text if len(token) > 1] for text in text_stem]
```

# Clustering Documents

* This part is computational expensive. I got the results once and store them. 

```python
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec

model = Word2Vec(text_stem, size=100, window=5, min_count=1, workers=4)
```

```python
def vectorize(list_of_docs, model):
    features = []
    for tokens in list_of_docs:
        zero_vector = np.zeros(model.vector_size)
        vectors = []
        for token in tokens:
            if token in model.wv:
                try:
                    vectors.append(model.wv[token])
                except KeyError:
                    continue
        if vectors:
            vectors = np.asarray(vectors)
            avg_vec = vectors.mean(axis=0)
            features.append(avg_vec)
        else:
            features.append(zero_vector)
    return features

vectorized_docs = vectorize(text_stem, model=model)
```

```python
def mbkmeans_clusters(X,k,mb,print_silhouette_values):
        km = MiniBatchKMeans(n_clusters=k, batch_size=mb).fit(X)
        print(f"For n_clusters = {k}")
        print(f"Silhouette coefficient: {silhouette_score(X, km.labels_):0.2f}")
        print(f"Inertia:{km.inertia_}")

        if print_silhouette_values:
            sample_silhouette_values = silhouette_samples(X, km.labels_)
            print(f"Silhouette values:")
            silhouette_values = []
            for i in range(k):
                cluster_silhouette_values = sample_silhouette_values[km.labels_ == i]
                silhouette_values.append(
                    (
                        i,
                        cluster_silhouette_values.shape[0],
                        cluster_silhouette_values.mean(),
                        cluster_silhouette_values.min(),
                        cluster_silhouette_values.max(),
                    )
                )
            silhouette_values = sorted(
                silhouette_values, key=lambda tup: tup[2], reverse=True
            )
            for s in silhouette_values:
                print(
                    f"    Cluster {s[0]}: Size:{s[1]} | Avg:{s[2]:.2f} | Min:{s[3]:.2f} | Max: {s[4]:.2f}"
                )
        return km, km.labels_
```

From here I am going to derive how we can group sales done by customers. The idea is find out patterns in the products based on their description to be able to define them as categories. Once, I define the category of my product proceed to analyze their behavior through the year and try to idenfy seasonality or posible impacts on promotions created by the marketing area. 

```python
# Definition use to get the cluster behavior.
# Find out clusters proportions over the half million purchases available on the dataset. 
# This helps to figure out which is the cluster with more items/less items. 
def get_clust_description(df):
    clust = np.transpose(df['cluster'].value_counts().index)
    freq = np.transpose(np.array(df['cluster'].value_counts()))
    df_freq = pd.DataFrame({"cluster" : clust, "frequency" : freq}).sort_values(by = ['frequency'], ascending=False)
    df_freq['percent'] = df_freq['frequency'] / df_freq['frequency'].sum() * 100
    df_freq_ = df_freq.head(10)
    df_freq_['percent_sum_top_n'] = df_freq_['percent'].sum()
    return df_freq_, df_freq.sort_values(by=['frequency'])
```

```python
# Definition created to substract nouns on the description of the product. Goal, once I do have my cluster
# which nouns are more popular on the cluster. Define categories.
def get_tokenize_noun(text):
    text_clean = text.split(', ')
    texts = []
    filters = ['NN','NNS','NNP']
    nouns = []
    for element in text_clean:
        if element is not '':
            texts.append(re.sub('[^a-zA-Z]+','',element))
    try:
        pos_texts = pos_tag(texts)
        for element in pos_texts:
            if element[1] in filters:
                nouns.append(element[0])
    except:
        nouns.append('noNoun')
    return nouns
```

```python
# Definition created to substract adjectives from the products descriptions. Goal, once I do have my cluster
# which adjectives are more popular on the cluster. Find out what characteristics the customers love the most/hate the most.
def get_tokenize_adj(text):
    text_clean = text.split(', ')
    texts = []
    filters = ['JJ','JJR','JJS']
    adj = []
    for element in text_clean:
        if element is not '':
            texts.append(re.sub('[^a-zA-Z]+','',element))
    try:
        pos_texts = pos_tag(texts)
        for element in pos_texts:
            if element[1] in filters:
                adj.append(element[0])
    except:
        adj.append('noAdj')
    return adj
```

```python
# Definition to create on the dataframe the adjective and noun tokenization.
def create_adj_nouns(df):
    df['tokenize_adj'] = list(map(get_tokenize_adj,df['text']))
    df['tokenize_noun'] = list(map(get_tokenize_noun,df['text']))
    return df
```

```python
# Dictionary to get the list of nouns or adjectives most popular on the cluster. 
# Use to find out a better description.
def get_dict_desc_top(pandas_column,n):
    noun_adj = list(itertools.chain(*pandas_column))
    noun_adj = pd.DataFrame({'adj': noun_adj})
    noun_adj = pd.DataFrame(noun_adj['adj'].value_counts())
    noun_adj = noun_adj.reset_index()
    noun_adj_dict = dict(noun_adj.head(n).values)
    return noun_adj_dict
```

```python
# Definition use to create the list of top 20 nouns and adjectives most popular.
def get_clust_desc(df):
    noun = get_dict_desc_top(df['tokenize_noun'],20)
    adj = get_dict_desc_top(df['tokenize_adj'],20)
    return noun, adj
```

```python
# Definition to get the tokens
def get_tokens(text):
    try:
        tokens = text.split(' ')
    except:
        tokens = 'no, Object'
    return tokens
```

```python
# Definition to show the most meaningful association rules between words to get a better description of the selected cluster.
def get_rules_properties(rules):
    df = pd.DataFrame(rules)
    df = df[df[2] != 1.0]
    df = df.sort_values(by = [2], ascending=False)
    return df
```

```python
# Definition to get the cluster properties. 
# Most popular nouns.
# Most popular adjectives.
# Most popular association between nouns-adjectives.
def get_cluster_properties(cluster_df, n_cluster, min_sup, min_conf):
    df = cluster_df[cluster_df['cluster']==n_cluster]
    df = create_adj_nouns(df)
    df['tokens'] = list(map(get_tokens,df['tokens']))
    df_noun, df_adj = get_clust_desc(df)
    freqItems, rules = apriori(df['tokens'], minSup=min_sup, minConf=min_conf)
    try:
        rules_df = get_rules_properties(rules)
    except:
        rules_df = 'no rules found'
    return df_noun, df_adj, df, rules_df
```

```python
def get_country_trends(country):
    important_clusters = [12,1,19,8,6,10]
    cluster_names = ['Baking Articles', 'Christmas Articles', 'Pantry/Tea Sets', 'Hangers, Wickers and Picnic Articles','Bags/Lunch Articles','Household Decoration']
    #10,8,6,19,1,12
    colors = ['brown','red','green','pink','black','orange'] 
    color = 'orange'
    plt.figure()
    for i in range(len(important_clusters)):
        df_country = df_countries_info[df_countries_info['Country'] == country]
        df_country = df_country[df_country['product_type'] == important_clusters[i]] 
        plt.plot(df_country['date'], df_country['Total_Price'], label = cluster_names[i], linestyle="-", color = colors[i])
    plt.legend()
    plt.xlabel('Date Time')
    plt.ylabel('Amount of Sales per Dollars')
    title = str('Sales Behavior ') + country
    plt.title(title)
    plt.show()
    pass
```
